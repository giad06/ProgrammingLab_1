## 1. Indicizzazione booleana
1. Creare un nuovo vettore con tutti i numeri primi tra 0 e 20. I numeri primi compresi tra 0 e 20 sono: 2, 3, 5, 7, 11, 13, 17, 19.

2. Utilizza un test logico (ad esempio >, <, ==, !=) per selezionare tutte le voci del vettore maggiori di 10.

3. Utilizza un test logico per selezionare tutti i numeri primi pari dalla lista.

```python
import numpy as np

def primi(n):
    if n<2:
        return False
    for i in range(2,n):
        if n%i == 0:
            return False
    return True

arr = np.array([i for i in range(2,20) if primi(i)])
print("Vettore con numeri primi:", arr)

print("Maggiori di 10:", arr[arr>10])

print("Primi e pari:", arr[arr%2==0])
```

## 3. Seleziona il numero più vicino a 0.5 in un array casuale
Genera un array 10 x 3 con numeri casuali nell'intervallo [0,1]. Per ogni riga, seleziona il numero più vicino a 0.5. Per farlo usa il fancy indexing.

```python
np.random.seed(0) # Fisso (avrò sempre stessi numeri)
A = np.random.rand(10,3) # Genero matrice 10x3 con valori casuali in [0,1], ogni riga 3 numeri
print(A,"\n")

B = np.abs(A-0.5) # creo nuovo array con valore assoluto della distanza di ogni elemento da 0.5
print(B,"\n")

index = np.argmin(B, axis=1) # mi restituisce indici di ogni colonna con distanza minima da 0.5 (per ogni riga)
print(index, "\n")
print("Array contenente i numeri piu vicini a 0.5 per ogni riga:")
A[np.arange(A.shape[0]), index] # fancy indexing:  A[righe, colonne]
```

# 5. Manipolazione di vettori
1. Creare un vettore contenente i seguenti stipendi: 50.000, 105.250, 55.000, 89.000.
Qual è il costo totale del personale (somma di tutti gli stipendi dell'azienda)?

2. Supponiamo che il nostro malvagio CEO abbia deciso di darsi un aumento.
Modifica il vettore degli stipendi in modo che il CEO, la persona che guadagna 105.250 dollari, riceva un aumento del 15%.

3. Il 115% di 105.250 dollari è 121.037,50 dollari.
Questo valore è presente nel tuo array? Se no, riesci a capire il motivo?

4. Ricrea il vettore e utilizza l'argomento dtype in modo che, quando il CEO riceve un aumento del 15%, il suo stipendio finale sia esattamente 121.037,50 dollari.

5. Questo aumento ha infastidito l'impiegata con lo stipendio più basso (50.000 dollari), che ora pretende un aumento del 20%.
Modifica il vettore per riflettere questo cambiamento.

6. L'aumento dato al CEO e all'impiegata ha irritato gli altri due dipendenti, quindi ora devono ricevere un aumento del 10%.
Modifica il vettore di conseguenza.

7. Calcola il costo totale del personale dopo tutti gli aumenti.
Alla fine, quanto è costato all'azienda l'aumento di circa 16.000 dollari del CEO?

```python
# 1
stipendi = np.array([50000, 105250, 55000, 89000])
print("Gli stipendi sono: ", stipendi)
costo_tot = stipendi.sum()
print(f"Il costo totale del personale è {costo_tot}$")

# 2
stipendi[1] *= 1.15 # Aumento CEO del 15%
print("Gli stipendi con l'aumento del CEO sono:", stipendi)

# 3
stipendi.dtype
# Il valore 121037,50 non è presente perchè il tipo degli elementi del vettore è int64 e io sto cercando un float

# 4
stipendi = np.array([50000, 105250, 55000, 89000], dtype=float)
stipendi[1] *= 1.15
print("Gli stipendi con l'aumento del CEO (float) sono:", stipendi)

# 5
stipendi[0] *= 1.20
print("Stipendi con aumento dell'impiegata infastidita:", stipendi)

# 6
stipendi[(stipendi != stipendi[0]) & (stipendi != stipendi[1])] *= 1.10
print("Tutti gli stipendi aumentati: ", stipendi)

# 7
finale = stipendi.sum()
print(f"Costo totale del personale dopo tutti gli aumenti: {finale}$")

costo_extra = finale - costo_tot
print(f"L'aumento totale è costato: {costo_extra}$")
```
