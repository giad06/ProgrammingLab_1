# 4. Rispondere alle domande
Trasforma le seguenti richiesti in formule per estrarre infromazioni dal dataframe
- Mostrami le vendite maggiori di 10

- Mostrami i dati del 2018

- Mostrami le vendite maggiori di 13 e l'anno è il 2018

- Mostrami tutto TRANNE i casi in cui le vendite sono maggiori di 13 e l'anno è il 2018

- Mostrami i dati dove le vendite divise per 3 sono maggiori di 3

- Mostrami i dipendenti i cui nomi sono alfabeticamente dopo la J



```python
import pandas as pd

sales = pd.DataFrame(
    data={
        "employee": [
            "Katrina",
            "Guanyu",
            "Jan",
            "Roman",
            "Jacqueline",
            "Paola",
            "Esperanza",
            "Alaina",
            "Egweyn",
        ],
        "sales": [14, 17, 6, 12, 8, 3, 7, 15, 5],
        "year": [2018, 2019, 2020, 2018, 2020, 2019, 2019, 2020, 2020],
    }
)
sales
```

```python
# Mostrami le vendite maggiori di 10
sales[sales['sales'] > 10]
```

```python
# Mostrami i dati del 2018
sales[sales['year']==2018]
```

```python
# Mostrami le vendite maggiori di 13 e l'anno è il 2018
sales[(sales['sales'] > 13) & (sales['year'] == 2018)]
```

```python
# Mostrami tutto TRANNE i casi in cui le vendite sono maggiori di 13 e l'anno è il 2018
sales[~((sales['sales'] > 13) & (sales['year'] == 2018))]
```

```python
# Mostrami i dati dove le vendite divise per 3 sono maggiori di 3
sales[(sales['sales']/3) > 3]
```

```python
# Mostrami i dipendenti i cui nomi sono alfabeticamente dopo la J
sales.query('employee > "J"')
```

# 5. Analisi di un dataset di video giochi
1. Quanti videogiochi sono stati pubblicati
2. Quali sono i generi più popolari? Mostralo con un bar plot
3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
4. Prova a riprodurre questo plot 




```python
# Import dei pacchetti
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Quanti videogiochi sono stati pubblicati
videogiochi = df.shape[0]
print(f"Sono stati pubblicati {videogiochi} videogiochi")
```

```python
# Quali sono i generi più popolari? Mostralo con un bar plot
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(genre_counts.index, genre_counts.values)
plt.title('Popolarità dei generi')
plt.xlabel('Genere')
plt.ylabel('Numero di giochi')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#  Plotta l'evoluzione del numero di giochi pubblicati nel tempo

#Pulisco
df_pulito = df.dropna(subset=["Year"])
df_pulito["Year"] = df_pulito["Year"].astype(int)

plt.figure(figsize=(12,6))
df_pulito.groupby("Year")["Name"].count().plot(marker="o")
plt.title("Evoluzione del numero di giochi pubblicati nel tempo")
plt.xlabel("Anno")
plt.ylabel("Numero di giochi")
plt.grid(True)
plt.show()

```

```python
sales_col = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']

# Raggruppo per genere e sommo vendite per ogni regione
genre_sales = df.groupby('Genre')[sales_col].sum()

# sommo le 4 regioni per riga (vendite totali per ogni genere), ordino dal piu grande e prendo primi 5
top5_gen = (
    genre_sales.sum(axis=1).sort_values(ascending=False).head(5).index
)

# Uso i top 5 per estrarli da principale
genre_sales = genre_sales.loc[top5_gen]

# Creazione plot
genre_sales.plot(kind='bar', figsize=(10, 7))
plt.title('Vendite per regione per genere')
plt.xlabel('Genre')
plt.ylabel('Milioni di unità')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
